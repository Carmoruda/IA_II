## PyTorch

```python
import torch # type: ignore
import torch.nn as nn  # type: ignore # módulo de redes neuronales
from torch.utils.data import Dataset, DataLoader, random_split # type: ignore # módulo de carga de datos
import torch.optim as optim # type: ignore # módulo de optimización
import numpy as np # módulo de manejo de arrays
```

```python
# Tensores básicos
# Creación de tensores
x = torch.tensor([1, 2, 3])  # Tensor 1D
matriz = torch.tensor([[1, 2], [3, 4]])  # Tensor 2D
random_tensor = torch.rand(3, 3)  # Tensor aleatorio 3x3
```

```python
# Operaciones básicas
suma = x + 5  # Suma elemento a elemento
multiplicacion = x * 2  # Multiplicación elemento a elemento
producto_escalar = torch.dot(x, x)  # Producto escalar
```

```python
# 1. Primero, creamos un dataset personalizado
class CustomDataset(Dataset):
    def __init__(self, num_samples=1000):
        # Generamos datos de ejemplo: x son números aleatorios y y = 2x + ruido
        self.x = torch.randn(num_samples, 1)
        self.y = 2 * self.x + torch.randn(num_samples, 1) * 0.1
        
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
```

```python
# 2. Creación de una red neuronal simple
class RedNeuronalSimple(nn.Module):
    def __init__(self):
        super(RedNeuronalSimple, self).__init__()
        # Capas de la red
        self.capa1 = nn.Linear(1, 5)  # Capa de entrada: 1 neuronas, salida: 5 neuronas
        self.capa2 = nn.Linear(5, 1)   # Capa de salida: 5 neuronas de entrada, 1 neuronas de salida
       # self.capa3 = nn.Linear(5, 10)   # Capa de salida: 5 neuronas de entrada, 1 neuronas de salida
       # self.capa4 = nn.Linear(10, 5)   # Capa de salida: 5 neuronas de entrada, 1 neuronas de salida
        self.relu = nn.ReLU()  # Función de activación ReLU
    
    def forward(self, x):
        # Definición del pase hacia adelante
        x = self.relu(self.capa1(x))
       # x = self.capa4(self.capa3(x))
        x = self.capa2(x)
        return x
```

```python
# 3. Entrenamiento básico

# Configuramos los hiperparámetros
batch_size = 32
learning_rate = 0.01
num_epochs = 500
validation_split = 0.2 

# Creamos el dataset y dataloader
dataset = CustomDataset()
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Creamos el dataset y lo dividimos en entrenamiento y validación
dataset = CustomDataset()
dataset_size = len(dataset)
validation_size = int(validation_split * dataset_size)
train_size = dataset_size - validation_size
train_dataset, val_dataset = random_split(dataset, [train_size, validation_size])


# Creamos los dataloaders para entrenamiento y validación
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Creamos instancia del modelo
model = RedNeuronalSimple()

# Definimos la función de pérdida y el optimizador
criterion = nn.MSELoss()  # Pérdida de error cuadrático medio
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate) #Descenso del gradiente estocástico

# Función para evaluar el modelo
def evaluate_model(model, dataloader, criterion):
    model.eval() # Modo de evaluación, deshabilita dropout y batch normalization
    total_loss = 0 
    with torch.no_grad(): # Deshabilitamos el cálculo de gradientes para que vaya más rápido
        for batch_x, batch_y in dataloader:
            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)
            total_loss += loss.item()
    return total_loss / len(dataloader)

#Entrenamiento y validación del modelo
train_losses = []
val_losses = []
best_val_loss = float('inf')
best_model_state = None

for epoch in range(num_epochs):
    # Modo entrenamiento
    model.train()
    train_loss = 0
    for batch_x, batch_y in train_loader:
        # Forward pass
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        
        # Backward pass y optimización
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
    
    # Calcular pérdida promedio de entrenamiento
    avg_train_loss = train_loss / len(train_loader)
    train_losses.append(avg_train_loss)
    
    # Validación
    val_loss = evaluate_model(model, val_loader, criterion)
    val_losses.append(val_loss)
    
    # Guardar el mejor modelo
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_model_state = model.state_dict().copy()
    
    # Imprimir progreso cada 10 épocas
    if (epoch + 1) % 10 == 0:
        print(f'Época [{epoch+1}/{num_epochs}]')
        print(f'Pérdida de entrenamiento: {avg_train_loss:.4f}')
        print(f'Pérdida de validación: {val_loss:.4f}')
        print('-' * 50)

# Cargar el mejor modelo
model.load_state_dict(best_model_state)
print(f'\nMejor pérdida de validación: {best_val_loss:.4f}')
```

```python
# 4. Evaluación final del modelo
model.eval()
with torch.no_grad():
    test_x = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
    predictions = model(test_x)
    print("\nPredicciones del mejor modelo:")
    for x, pred in zip(test_x, predictions):
        print(f"Entrada: {x.item():.1f}, Predicción: {pred.item():.4f}")
```

```python

# 5. Visualización de las pérdidas de entrenamiento y validación
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Entrenamiento')
plt.plot(val_losses, label='Validación')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.title('Pérdida de Entrenamiento vs Validación')
plt.legend()
plt.grid(True)
plt.show()
```
